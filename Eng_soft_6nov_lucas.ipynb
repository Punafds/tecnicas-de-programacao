```python
print('Digite a primeira nota:')
nota1 = float(input())

print('Digite a segunda nota:')
nota2 = float(input())

print('Digite a terceira nota:')
nota3 = float(input())

media_ponderada = (nota1 + nota2 + nota3) / 3

print(f'A média ponderada é: {media_ponderada}')
```

```python
print('Digite a primeira nota:')
nota1 = float(input())
print('Digite o peso da primeira nota:')
peso1 = float(input())

print('Digite a segunda nota:')
nota2 = float(input())
print('Digite o peso da segunda nota:')
peso2 = float(input())

print('Digite a terceira nota:')
nota3 = float(input())
print('Digite o peso da terceira nota:')
peso3 = float(input())

media_ponderada = (nota1 * peso1 + nota2 * peso2 + nota3 * peso3) / (peso1 + peso2 + peso3)

print(f'A média ponderada é: {media_ponderada}')
```

```python
print('Digite um número:')
numero = input()
soma = 0
for digito in numero:
  soma += int(digito)

print(f'A soma dos dígitos é: {soma}')
```
